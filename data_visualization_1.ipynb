```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
titanic.isnull().sum()
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
```

```python
titanic.isnull().sum()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title("Count vs Age")
sns.histplot(data=titanic, x='age')
plt.subplot(1, 2, 2)
plt.title("Count vs Fare")
sns.histplot(data=titanic, x="fare", bins=20)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')

plt.subplot(2,2,1)
sns.histplot(data=titanic, x='age', hue='sex', multiple="dodge")

plt.subplot(2,2,2)
sns.histplot(data=titanic, x='age', hue='survived', multiple="dodge", shrink=.8)

plt.subplot(2,2,3)
sns.histplot(data=titanic, x='fare', hue='sex', multiple="dodge", binwidth=20, bins=10)

plt.subplot(2,2,4)
sns.histplot(data=titanic, x='fare', hue='survived', multiple="dodge", binwidth=20, bins=10)

plt.tight_layout()

plt.show()
```
